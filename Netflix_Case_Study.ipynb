#### Q7.A popular film means it is released at Netflix platforms within 1 year.Find all Films which were released in Netflix within 2 years of actual release.

```python
df["Release Year"] = pd.to_datetime(df["Release Date"], errors='coerce').dt.year
df["Netflix Release Year"] = pd.to_datetime(df["Netflix Release Date"],errors='coerce').dt.year

popular_films = df[
    (df["Series or Movie"] == "Movie") &
    (df["Netflix Release Year"].notna()) &
    (df["Release Year"].notna()) &
    ((df["Netflix Release Year"] - df["Release Year"]) <= 2)
]

# Select relevant columns
print(popular_films[["Title", "Release Year", "Netflix Release Year"]])
```
